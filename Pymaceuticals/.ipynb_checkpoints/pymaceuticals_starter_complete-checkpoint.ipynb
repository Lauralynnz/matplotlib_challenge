```python
# THREE OBSERVATIONS ABOUT THE RESULTS OF THE STUDY

# 1. Tumor volume decreased over time for the treatments of Capomulin and Ramicane. The volume increased over time for everything else.

# 2. Metastatic response to all treatments increased over time. Meaning, even though the tumor volume decreased for a couple of the treatments, the tumors continued to spread with all the treatments.

# 3. Capomulin and Ramicane had the highest survival rate, but if the tumors are continuing to spread even during treatment we donâ€™t know the study has run long enough to tell if these two treatments are actually saving patients.
```

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset on Mouse ID
#print(mouse_data_df)
#print(trial_data_df)

combined_data_complete = pd.merge(clinical_data, mouse_data, how="left", on=["Mouse ID", "Mouse ID"])


# Display the data table for preview
combined_data_complete

# Change Index to Mouse ID (remove line count on left)

combined_data_complete = combined_data_complete.set_index("Mouse ID")
combined_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

mean_tumor_vol = combined_data_complete.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
mean_tumor_vol


# Convert to DataFrame
#EXAMPLE: Converting a GroupBy object into a DataFrame
#grouped_tumor_volume = pd.DataFrame(
#    mean_tumor_volume["duration (seconds)"].mean())
#print(grouped_tumor_volume)

mean_tumor_vol = pd.DataFrame(mean_tumor_vol)
mean_tumor_vol

#EXAMPLE: Reset the index for this DataFrame so "Assignee" is a column
#user_bug_percent = user_bug_percent.reset_index()
#user_bug_percent.head()

mean_tumor_vol = mean_tumor_vol.reset_index()

# Preview DataFrame

mean_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

sem_tumor_vols = combined_data_complete.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]
sem_tumor_vols


# Convert to DataFrame

sem_tumor_vols = pd.DataFrame(sem_tumor_vols)
sem_tumor_vols

sem_tumor_vols = sem_tumor_vols.reset_index()


# Preview DataFrame

sem_tumor_vols.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mean_tumor_vol = mean_tumor_vol.reset_index()
#print(mean_tumor_vol)

mean_tumor_vol_pivot = mean_tumor_vol.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
#print(mean_tumor_vol_pivot)

sem_tumor_vols = sem_tumor_vols.reset_index()
#print (sem_tumor_vols)

sem_tumor_vols_pivot = sem_tumor_vols.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
#print(sem_tumor_vols_pivot)

# Preview that Reformatting worked
mean_tumor_vol_pivot.head()



```

```python
#NOTE: COMPLETE THIS WHEN ABOVE FRAME IS DEBUGGED
plt.figure(figsize=(20,20))

# Generate the Plot (Accounting for percentages)
plt.plot(100 * num_mice_survive_pivot["Capomulin"] / 25, "gold", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Ceftamin"] / 25, "blue", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Infubinol"] / 25, "green", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Ketapril"] / 25, "pink", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Naftisol"] / 25, "slategrey", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Placebo"] / 25, "orange", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Propriva"] / 25, "lime", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Ramicane"] / 25, "maroon", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Stelasyn"] / 25, "cyan", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)
plt.plot(100 * num_mice_survive_pivot["Zoniferol"] / 25, "magenta", linestyle="dashed", marker="o", markersize=5, linewidth=1.50)


# Save the Figure

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(["Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva", "Ramicane", "Stelasyn", "Zoniferol"], 
               loc="best", fontsize="small", fancybox=True)


# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_pct_change = 100 * (mean_tumor_vol_pivot.iloc[-1] - mean_tumor_vol_pivot.iloc[0])/ mean_tumor_vol_pivot.iloc[0]
tumor_pct_change_sem = 100 * (sem_tumor_vols_pivot.iloc[-1] - sem_tumor_vols_pivot.iloc[0])/ sem_tumor_vols_pivot.iloc[0]

# Display the data to confirm
tumor_pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_changes = (tumor_pct_change["Capomulin"],
               tumor_pct_change["Infubinol"],
               tumor_pct_change["Ketapril"],
               tumor_pct_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_changes))
width = 1
rectsPass = ax.bar(ind[0], pct_changes[0], width, color="blue")
rectsFail = ax.bar(ind[1:], pct_changes[1:], width, color="pink")

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 day Treatment")
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(("Capomulin", "Infubinol", "Ketapril", "Placebo"))
ax.set_autoscaley_on(False)
ax.set_ylim([-30, 70])
ax.grid(True)


# Use functions to label the percentages of changes
def autolabelFail(rects):
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width() / 2. , 3, "%d%%" % int(height),
                   ha="center", va="bottom", color="white")

# Call functions to implement the function calls
def autolabelPass(rects):
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width() / 2. , -8, "%d%%" % int(height),
                   ha="center", va="bottom", color="white")


# Save the Figure
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Show the Figure
fig.show()

```
